```python
def get_data():
    # Get the current working directory
    base_path = os.getcwd()
    # Construct the full path to the CSV file
    file_path = os.path.join(base_path, 'data', 'mock_user_picks.csv')
    
    # Check if the file exists at the constructed path
    if not os.path.exists(file_path):
        # If not, adjust the path assuming the script is run from the notebooks directory
        file_path = os.path.join(base_path, '..', 'data', 'mock_user_picks.csv')
    
    # Read the CSV file into a DataFrame
    data = pd.read_csv(file_path)
    return data 

# Read the CSV file into a DataFrame
df1 = get_data()

# Display the first few rows of the DataFrame to verify
print(df1.head())
df1.info()



# Calculate the implied probability from the odds
df1['Implied_Probability'] = 1 / df1['Odds']

# Calculate the expected value (EV) of each bet
df1['Expected_Value'] = (df1['Implied_Probability'] * df1['Payout']) - ((1 - df1['Implied_Probability']) * df1['Bet_Amount'])

# Calculate the total expected value and total actual payouts
total_expected_value = df1['Expected_Value'].sum()
total_actual_payouts = df1['Payout'].sum()

# Display the first few rows of the DataFrame to verify
print(df1.head())
df1.info()

# Print the total expected value and total actual payouts
print(f"Total Expected Value: {total_expected_value}")
print(f"Total Actual Payouts: {total_actual_payouts}")
```

```python
# Assuming df1 is your DataFrame with the calculated Expected_Value
# Group by User_ID and aggregate the data
user_summary = df1.groupby('User_ID').agg(
    Total_Expected_Value=('Expected_Value', 'sum')
).reset_index()

# Create bins for the expected values
bins = 10
user_summary['EV_Bins'] = pd.cut(user_summary['Total_Expected_Value'], bins=bins)

# Count the number of users in each bin
ev_bin_counts = user_summary['EV_Bins'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=ev_bin_counts.index.astype(str), y=ev_bin_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Distribution of Total Expected Values for Users')
plt.xlabel('Total Expected Value Bins')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Assuming df1 is your DataFrame with the calculated Profit
# Group by User_ID and aggregate the data
user_summary = df1.groupby('User_ID').agg(
    Total_Profit=('Profit', 'sum')
).reset_index()

# Create bins for the profits
bins = 10
user_summary['Profit_Bins'] = pd.cut(user_summary['Total_Profit'], bins=bins)

# Count the number of users in each bin
profit_bin_counts = user_summary['Profit_Bins'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_bin_counts.index.astype(str), y=profit_bin_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Distribution of Total Profits for Users')
plt.xlabel('Total Profit Bins')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Group by User_ID and aggregate the data
user_summary = df1.groupby('User_ID').agg(
    Total_Expected_Value=('Expected_Value', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

# Create bins for the expected values and profits
bins = 10
user_summary['EV_Bins'] = pd.cut(user_summary['Total_Expected_Value'], bins=bins)
user_summary['Profit_Bins'] = pd.cut(user_summary['Total_Profit'], bins=bins)

# Count the number of users in each bin for expected values and profits
ev_bin_counts = user_summary['EV_Bins'].value_counts().sort_index()
profit_bin_counts = user_summary['Profit_Bins'].value_counts().sort_index()

# Plot the side-by-side bar charts
fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# Bar chart for expected values
sns.barplot(x=ev_bin_counts.index.astype(str), y=ev_bin_counts.values, palette='viridis', ax=axes[0])
axes[0].set_title('Distribution of Total Expected Values for Users')
axes[0].set_xlabel('Total Expected Value Bins')
axes[0].set_ylabel('Number of Users')
axes[0].tick_params(axis='x', rotation=45)

# Bar chart for profits
sns.barplot(x=profit_bin_counts.index.astype(str), y=profit_bin_counts.values, palette='viridis', ax=axes[1])
axes[1].set_title('Distribution of Total Profits for Users')
axes[1].set_xlabel('Total Profit Bins')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
